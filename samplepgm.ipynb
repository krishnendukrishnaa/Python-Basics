```python
#largest number among the three input numbers
def largest():
  a=int(input("Enter the 1st number: "))
  b=int(input("Enter the 2nd number: "))
  c=int(input("Enter the 3rd number: "))
  if((a>b)and(a>c)):
    print(a)
  elif((b>a)and(b>c)):
    print(b)
  else:
    print(c)
largest()
```
