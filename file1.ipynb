```python
def length():
  a=input("Enter the word: ")
  b=input("Enter the next word: ")
  print(a,len(a))
  print(b,len(b))
  if(len(a)==len(b)):
    print(a)
    print(b)
length()
```

```python
def even():
  a=int(input("Enter the value:"))
  if(a%2==0):
    print("It is an even number")
  else:
    print("It is an odd number")
even()
```

```python
def dictionary():
  a=int(input("Enter the limit:"))
  d=dict()
  for x in range(1,a+1):
    d[x]=x**2
  print(d)
dictionary()
```

```python
def square():
  d=dict()
  for x in range(1,20):
    d[x]=x**2
  print(d)
square()
```

```python
def sq():
  d=dict()
  for x in range(1,20):
    d[x]=x**2
    print(d[x])
  print(d)
sq()
```
