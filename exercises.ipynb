```python
class bmicalc:
  def __init__(self,w,h):
    self.w=w
    self.h=h
  def bmicalculation(self):
    self.bmi=(self.w/self.h/self.h)*10000
    print(self.bmi)
  def display(self):
    if(self.bmi<=18.5):
      print("Under weight.")
    elif(self.bmi>=18.5)or(self.bmi<=24.9):
      print("Normal weight")
    elif(self.bmi>=25)or(self.bmi<=29.9):
      print("Over weight")
    else:
      print("Obese")
o=bmicalc(56,159)
o.bmicalculation()
o.display()
```
