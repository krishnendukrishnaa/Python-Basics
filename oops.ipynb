```python
class person:
  def __init__(self,name,age):
    self.name=name
    self.age=age
  def display(self):
    print(self.name,self.age)
  def place(self,location):
    self.location=location
    print(self.location)
obj=person("Ram",24)
obj.display()
obj.place("Kochi")
```

```python
#same value for every class instance
class vehicle:
  def __init__(self,color,price):
    self.color=color
    self.price=price
  def display(self):
    print(self.color,self.price)
class bus(vehicle):
  def busroot(self):
    print("Bus root ")
class car(bus):
  def brand(self):
    print("The brand is Benze")
a=car("White",20000)
a.display()
a.busroot()
a.brand()
```

```python
class animal:
  def __init__(self,a):
    self.a=a
  def display(self):
    print(self.a)
class dog(animal):
  def parent(self):
    print("The parent")
class dogchild(dog):
  def child(self):
    print("Child of the dog")
a=dogchild("Husky")
a.display()
a.parent()
a.child()
```

```python
class calculation1:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  def add(self):
    print("Add:",self.a+self.b)
class calculation2:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  def sub(self):
    print("Sub:",self.a-self.b)
class calculation3(calculation1,calculation2):
  def mul(self):
    print("Multiplication:",self.a*self.b)
a=int(input("Enter the value: "))
b=int(input("Ente the next value: "))
c=calculation3(a,b)
c.add()
c.sub()
c.mul()
```
