```python
#multiplication table
def multbl():
  a=int(input("Enter the Number: "))
  for i in range(1,11):
    b=a*i
    print(a,'x',i,'=',b)
multbl()
```

```python
#Area of rectangle
class area:
  def __init__(self,h,w):
    self.h=h
    self.w=w
  def display(self):
    print("Area of a Ractangle: ", self.h*self.w)
h=int(input("Enter the Hieght:"))
w=int(input("Enter the width: "))
o=area(h,w)
o.display()
```

```python
#string is palindrome
a=input("Enter the word: ")
if(a==a[::-1]):
  print("The stirng is palindrome")
else:
  print("The stirng is not palindrome")
```

```python
#number is palindrome
a=int(input("Enter the number: "))
r=0
while(a>0):
  b=a%10
  r=r*10+b
  a=a//10
if(a==r):
    print("The number is palindrome")
else:
    print("The number is not palindrome")
```

```python
#sum of the digit
a=int(input("Enter the number: "))
sum=0
while(a!=0):
  sum=sum+(a%10)
  a=a//10
print(sum)
```

```python
#reverse of a number
a=int(input("Enter the number: "))
rev=0
while(a>0):
  r=a%10
  rev=(rev*10)+r
  a=a//10
print(rev)


```
