```python
def largest(a,b,c):
  if((a>b)and(a>c)):
    print(a)
  elif((b>a)and(b>c)):
    print(b)
  else:
    print(c)
largest(2,6,18)

```

```python
def zero(a):
  if(a>0):
    print("Posetive")
  elif(a<0):
    print("Negative")
  else:
    print("Zero")
zero(-7)
```

```python
def odd(a):
  if(a%2==0):
    print("Even")
  else:
    print("Odd")
odd(15)
```

```python
def avg(a,b,c,d,e):
  avg=(a+b+c+d+e)/5
  print(avg)
  if((avg>=90)and(avg<=100)):
    print("A+")
  elif((avg>=80)and(avg<=90)):
    print("A")
  elif((avg>=70)and(avg<=80)):
    print("B+")
  elif((avg>=60)and(avg<=70)):
    print("B")
  elif((avg>=50)and(avg<=60)):
    print("C+")
  elif((avg>=40)and(avg<=50)):
    print("C")
  else:
    print("Failed")
avg(80,80,90,90,80)
```

```python
def smallest(a,b,c):
  if((a<b)and(a<c)):
    print(a)
  elif((b<a)and(b<c)):
    print(b)
  else:
    print(c)
smallest(2,1,0)
```
